```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Topology graph
G = nx.read_graphml('graph/01-topology.graphml')

pos = {id: [data['x'], data['y']] for id, data in G.nodes(data=True)}

plt.figure(figsize=(6, 6))
nx.draw(G, pos=pos, node_size=10, width=0.5, arrowsize=5)
plt.show()
```

```python
# Graph with lane changes
G = nx.read_graphml('graph/03-long-edges.graphml')
P = nx.read_graphml('graph/04-topological-path.graphml')

plt.figure(figsize=(6, 6))
nx.draw(G, pos={id: [data['x'], data['y']] for id, data in G.nodes(data=True)}, node_size=10, width=0.5, arrowsize=5)
nx.draw(P, pos={id: [data['x'], data['y']] for id, data in P.nodes(data=True)}, node_size=10, width=2, node_color='r', edge_color='r', arrowsize=5)
plt.show()
```
